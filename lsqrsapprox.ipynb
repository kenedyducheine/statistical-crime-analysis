```python
import numpy as np 
import pandas as pd
from google.cloud import bigquery

#client_id = bigquery.Client("crime-analysis-491817")
from google.oauth2 import service_account

with open("credentials.txt", "r") as f:
    creds = f.read()
credentials = service_account.Credentials.from_service_account_file(creds)

client_id = bigquery.Client(project="crime-analysis-491817", credentials=credentials)

```

```python
query = """
SELECT * 
FROM `bigquery-public-data.chicago_crime.crime`
LIMIT 100
"""

df = client_id.query(query).to_dataframe()
print(df.head())
print(df.columns)
```
